Exercise 4.9, p. 152 (Temperature Conversion) Implement a `fahrenheit` function that returns the Fahrenheit equivalent of a Celsius temperature. Use the following formula:<br />
F = (9 / 5) * C + 32<br />
Use this function to print a chart showing the Fahrenheit equivalents of all Celsius temperatures in the range 0â€“100 degrees. Use one digit of precision for the results. Print the outputs in a neat tabular format.

```python
#defines range for function to operate
celsius = range(0,101) 

#convert numbers in range to fahrenheit
def fahrenheit(celsius):
    return (9 / 5) * celsius + 32

print("Celsius\tFahrenheit\n------------------")

#prints range and calculated values
for i in celsius:
    print(f'{i:>4}{fahrenheit(i):10.1f}')
```
