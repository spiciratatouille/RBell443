Perform K-Fold Cross-Validation with 10 splits and a `random_state` of 15. Calculate and print the accuracy scores for each fold and their overall mean accuracy. (5 pts)

```python
%matplotlib inline
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
GNB = GaussianNB()

breast_cancer = load_breast_cancer()

# Put remaining code here in sections numbered as directed above

# 1
print(breast_cancer.DESCR)
# 2
print(f'Samples and features per sample: \n {breast_cancer.data.shape}')
# 3
print(f'Targets: \n {breast_cancer.target.shape}')
# 4
print('Every 10th Label:')
print(breast_cancer.target[::10])
# 5
X_train, X_test, y_train, y_test = train_test_split(breast_cancer.data, breast_cancer.target, random_state = 15)
print('Training:\n', X_train.shape)
print('Test:\n', X_test.shape)
# 6
GNB.fit(X = X_train, y = y_train)
# 7
predicted = GNB.predict(X=X_test)
expected = y_test
missed = [(p,e) for (p,e) in zip(predicted, expected) if p!= e]
print('Wrong Predictions:\n',missed)
# 8
score = (GNB.score(X_test,y_test))*100
print(f' Score: {score:.2f}%')
# 9
confusion = confusion_matrix(y_true = expected, y_pred = predicted)
print('Confusion Matrix:\n',confusion)
# 10
names = [str(digit) for digit in breast_cancer.target_names]
print(classification_report(expected, predicted, target_names=names))
# Extra Credit
```
